## 3. ***Data Wrangling***

### Data Wrangling Code

```python
df['Timestamp IST'] = pd.to_datetime(df['Timestamp IST'], format='%d-%m-%Y %H:%M')
# Convert 'Timestamp' to datetime, assuming it's a Unix timestamp in milliseconds
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='ms')

# Check the unique values in 'Direction' column
print("Unique values in 'Direction' column before cleaning:", df['Direction'].unique())

# Standardize 'Direction' column to 'Buy' or 'Sell'
# Assuming any value containing 'Buy' is 'Buy' and 'Sell' is 'Sell', and others might be cleaned or investigated.
df['Direction'] = df['Direction'].apply(lambda x: 'Buy' if 'Buy' in x else ('Sell' if 'Sell' in x else x))

# Check the unique values in 'Direction' column after cleaning
print("Unique values in 'Direction' column after cleaning:", df['Direction'].unique())







```

```python
# Display info to verify data type changes
df.info()
```

```python
# Extract time-based features from 'Timestamp IST'
df['Year'] = df['Timestamp IST'].dt.year
df['Month'] = df['Timestamp IST'].dt.month
df['Day'] = df['Timestamp IST'].dt.day
df['DayOfWeek'] = df['Timestamp IST'].dt.dayofweek
df['Hour'] = df['Timestamp IST'].dt.hour
```

```python
# Convert 'Trade ID' to object type if it's an identifier
df['Trade ID'] = df['Trade ID'].astype(object)
```

```python
print("\nDataFrame after extracting time features and converting Trade ID type:")
df.info()
```

```python
df['Direction'].value_counts()
```

```python
# Verify the consistency between 'Side' and 'Direction'
# They both appear to be 'Buy' or 'Sell', let's check for any discrepancies.
print("\nAre 'Side' and 'Direction' consistent?")
print(df[df['Side'].str.lower() != df['Direction'].str.lower()].empty)
```

### What all manipulations have you done and insights you found?


Here's a summary of the data manipulations I've performed and the insights gathered from the wrangling process:

### Manipulations Done:

1.  **Timestamp Conversion**: The `Timestamp IST` column has been successfully converted from an object type to a `datetime64[ns]` format, using the specified `'%d-%m-%Y %H:%M'` format. The `Timestamp` column, initially a float, has also been converted to `datetime64[ns]` by interpreting its values as Unix milliseconds, resolving the `OutOfBoundsDatetime` error from the previous attempt.
2.  **Direction Column Standardisation (Partial)**: An attempt was made to standardize the `Direction` column to simply 'Buy' or 'Sell'. However, as observed from the unique values *after* cleaning, the logic applied (`lambda x: 'Buy' if 'Buy' in x else ('Sell' if 'Sell' in x else x)`) only changed values that explicitly contained 'Buy' or 'Sell'. This means more specific trade directions like 'Open Long', 'Close Long', 'Open Short', 'Close Short', etc., are still present. This indicates that the `Direction` column provides more granular information than a simple 'Buy' or 'Sell'.
3.  **Time-Based Feature Extraction**: New columns (`Year`, `Month`, `Day`, `DayOfWeek`, `Hour`) have been successfully extracted from the `Timestamp IST` column. These new features will be valuable for analyzing temporal patterns in the trading data.
4.  **Trade ID Type Conversion**: The `Trade ID` column, which was a float, has been converted to an `object` type, acknowledging its role as a unique identifier rather than a numerical value for calculation.

### Insights Found:

*   **Data Type Readiness**: The key timestamp columns are now correctly formatted as datetime objects, enabling proper time-series analysis.
*   **Direction vs. Side Inconsistency**: The check for consistency between `Side` and `Direction` returned `False`, confirming that these two columns are not entirely redundant. The `Direction` column retains more detailed information about the nature of the trade (e.g., whether it's opening or closing a position) beyond just a simple 'Buy' or 'Sell', which is captured by the `Side` column. This granularity in `Direction` could be important for deeper analysis of trading strategies.
*   **New Temporal Dimensions**: The extracted time-based features provide new dimensions to explore the dataset, such as trade volume by hour, day of the week, or month, which can reveal trading patterns or market behavior over time.
